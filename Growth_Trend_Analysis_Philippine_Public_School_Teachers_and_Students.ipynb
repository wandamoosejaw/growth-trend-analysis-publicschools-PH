# Data Reshaping/ Transformation

Upon examining/inspecting each dataset, we've discovered that the teachers_df having 13 columns is much more than the students' datasets which only have 4. The info is listed again below for reference.

As a solution, Pandas' melt() function will be utilized to convert the
 ```teachers_df ```
 from wide format into along format, matching the structure of the students' data.



````python
#use melt() to reshape the data frame
melted_teachers_df = teachers_df.melt(
                    id_vars=['Region', 'Level'], value_vars=['SY 2010-2011', 'SY 2011-2012', 'SY 2012-2013', 'SY 2013-2014',
                                                             'SY 2014-2015', 'SY 2015-2016', 'SY 2016-2017', 'SY 2017-2018',
                                                             'SY 2018-2019', 'SY 2019-2020', 'SY 2020-2021'],
                    var_name='School Year',
                    value_name='Number of Teachers'
)
melted_teachers_df.head(5)
````

````python
# Standardize the 'School Year' format for the teachers DataFrame
melted_teachers_df['School Year'] = melted_teachers_df['School Year'].str.replace('SY ', '', regex=False)

# Standardize the 'School Year' format for the senior high DataFrame
senior_high_df['School Year'] = senior_high_df['School Year'].str.replace('SY ', '', regex=False)

````

For the last step, the  ```Grade Level``` column for students datasets are specified compared to the teacher's data wherein the level are generalized to elementary, junior high school, and others. So we need to categorize the students too in a general manner.



````python
# Add a 'School Level' column to the elementary student DataFrame
elementary_df['School Level'] = 'Elementary'

# Add a 'School Level' column to the junior high student DataFrame
junior_high_df['School Level'] = 'Junior High'

# Add a 'School Level' column to the senior high student DataFrame
senior_high_df['School Level'] = 'Senior High'
````

````python
# Combine the students' datasets
student_dfs = [elementary_df, junior_high_df, senior_high_df]

# Use pd.concat() to stack them vertically
students_df = pd.concat(student_dfs, ignore_index=True)

# The axis=1 argument tells pandas to drop a column (not a row)
# The inplace=True argument modifies the dataframe directly
students_df.drop('Grade Level', axis=1, inplace=True)

# Inspect the data
students_df.info()
````

````python
students_df.head(5)
````

Merge the students_df with melted_teachers_df

````python
# Rename the 'Level' column in the melted teachers DataFrame to match the students DataFrame
melted_teachers_df.rename(columns={'Level': 'School Level'}, inplace=True)

# Standardize the names in the junior high dataframe
melted_teachers_df['School Level'] = melted_teachers_df['School Level'].replace('Junior High School', 'Junior High')

# Standardize the names in the senior high dataframe
melted_teachers_df['School Level'] = melted_teachers_df['School Level'].replace('Senior High School', 'Senior High')

melted_teachers_df.head(100)
````

````python
# Merge the students and teachers DataFrames on the common columns
# Using an 'outer' merge to ensure we don't lose any data
final_df = pd.merge(students_df, melted_teachers_df, on=['Region', 'School Year', 'School Level'], how='outer')

# Now check the final merged DataFrame
print(final_df.info())
print(final_df.head())
````

````python
# Fill NaN values in 'Students' and 'Number of Teachers' columns with 0
final_df['Students'].fillna(0, inplace=True)
final_df['Number of Teachers'].fillna(0, inplace=True)

# Check the result to confirm there are no more NaNs in these columns
print(final_df.info())
````

# **Analysis Phase**

Since the data is cleaned and preprocessed, it's fitting to begin the analysis.

````python
#Calculate the student-teacher ratio
final_df['Teachers_per_Student_Ratio'] = final_df['Number of Teachers'] / final_df['Students']
````

````python
# Group by School Year and School Level and calculate the sums
summary_df = final_df.groupby(['School Year', 'School Level']).agg(
    Total_Students=('Students', 'sum'),
    Total_Teachers=('Number of Teachers', 'sum')
).reset_index()

# Calculate the ratio on this summary data as well
summary_df['Total_Teachers_per_Student'] = summary_df['Total_Teachers'] / summary_df['Total_Students']
````

````python
summary_df.head(10)
````

````python
# Standardize the names in the junior high dataframe
junior_high_df['School Level'] = junior_high_df['School Level'].replace('Junior High School', 'Junior High')

# Standardize the names in the senior high dataframe
senior_high_df['School Level'] = senior_high_df['School Level'].replace('Senior High School', 'Senior High')
````

````python
# Replace all 'inf' values with 0
final_df.replace([np.inf, -np.inf], 0, inplace=True)

# Replace any remaining NaN values with 0
final_df['Teachers_per_Student_Ratio'].fillna(0, inplace=True)

# Check the result
print(final_df.describe())
````

Visualize the computation

````python
# Now, create the line plot
plt.figure(figsize=(10,8))
sns.lineplot(data=summary_df, x='School Year', y='Total_Students', hue='School Level')
plt.title('Total Students by School Level Over Time')
plt.ylabel('Total Students')
plt.xticks(rotation=45)
plt.show()
````

````python
# Aggregate your data to get the summary
summary_df = final_df.groupby(['School Year', 'School Level']).agg(
    Total_Students=('Number of Students', 'sum'),
    Total_Teachers=('Number of Teachers', 'sum')
).reset_index()

# Calculate the overall ratio
summary_df['Overall_Ratio'] = summary_df['Total_Teachers'] / summary_df['Total_Students']

# Now, create two separate plots

# Plot 1: Total Student Population Growth
plt.figure(figsize=(10, 6))
sns.lineplot(data=summary_df, x='School Year', y='Total_Students', hue='School Level')
plt.title('Total Students by School Level Over Time')
plt.ylabel('Total Students')
plt.xticks(rotation=45)
plt.show()

# Plot 2: Ratio Trend
plt.figure(figsize=(10, 6))
sns.lineplot(data=summary_df, x='School Year', y='Overall_Ratio', hue='School Level')
plt.title('Teacher-Student Ratio by School Level Over Time')
plt.ylabel('Teachers per Student')
plt.xticks(rotation=45)
plt.show()
````
