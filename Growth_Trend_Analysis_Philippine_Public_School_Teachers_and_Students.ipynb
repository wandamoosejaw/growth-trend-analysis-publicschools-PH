As a solution, Pandas' melt() function will be utilized to convert the
 ```teachers_df ```
 from wide format into along format, matching the structure of the students' data.



````python
#use melt() to reshape the data frame
melted_teachers_df = teachers_df.melt(
                    id_vars=['Region', 'Level'], value_vars=['SY 2010-2011', 'SY 2011-2012', 'SY 2012-2013', 'SY 2013-2014',
                                                             'SY 2014-2015', 'SY 2015-2016', 'SY 2016-2017', 'SY 2017-2018',
                                                             'SY 2018-2019', 'SY 2019-2020', 'SY 2020-2021'],
                    var_name='School Year',
                    value_name='Number of Teachers'
)
melted_teachers_df.head(5)
````

For the last step, the  ```Grade Level``` column for students datasets are specified compared to the teacher's data wherein the level are generalized to elementary, junior high school, and others. So we need to categorize the students too in a general manner.



````python
# Add a 'School Level' column to the elementary student DataFrame
elementary_df['School Level'] = 'Elementary'

# Add a 'School Level' column to the junior high student DataFrame
junior_high_df['School Level'] = 'Junior High'

# Add a 'School Level' column to the senior high student DataFrame
senior_high_df['School Level'] = 'Senior High'
````

````python
# Combine the students' datasets
student_dfs = [elementary_df, junior_high_df, senior_high_df]

# Use pd.concat() to stack them vertically
students_df = pd.concat(student_dfs, ignore_index=True)

# The axis=1 argument tells pandas to drop a column (not a row)
# The inplace=True argument modifies the dataframe directly
students_df.drop('Grade Level', axis=1, inplace=True)

# Inspect the data
students_df.info()
````

````python
# Fill NaN values in 'Students' and 'Number of Teachers' columns with 0
final_df['Students'].fillna(0, inplace=True)
final_df['Number of Teachers'].fillna(0, inplace=True)

# Check the result to confirm there are no more NaNs in these columns
print(final_df.info())
````

# **Analysis Phase**

Since the data is cleaned and preprocessed, it's fitting to begin the analysis.

````python
#Calculate the student-teacher ratio
final_df['Teachers_per_Student_Ratio'] = final_df['Number of Teachers'] / final_df['Students']
````

````python
# Group by School Year and School Level and calculate the sums
summary_df = final_df.groupby(['School Year', 'School Level']).agg(
    Total_Students=('Students', 'sum'),
    Total_Teachers=('Number of Teachers', 'sum')
).reset_index()

# Calculate the ratio on this summary data as well
summary_df['Total_Teachers_per_Student'] = summary_df['Total_Teachers'] / summary_df['Total_Students']
````
